```python
import numpy as np
```

```python
seq={}
with open("btub_haplo.txt",'r') as b:
    for line in b:
        line = line.strip().split("\t")
        seq[line[0]] = line[1]
```

```python
import pandas as pd

mat = np.zeros(shape=(len(seq.keys()),len(seq.keys())))
print(mat.shape)
n=0
m=0
for i in seq.keys():
    m=0
    for k in seq.keys():        
        mat[n,m] = sum(abs(int(a)-int(b)) for a, b in zip(seq[i], seq[k])) ## No. allelic differences betw. seq.
        m+=1
    n+=1


##-- Output Dxy between phased genotypes

df = pd.DataFrame(mat)
df.columns = seq.keys()
#df.rows = seq.keys()
df.to_csv("haplo_aldiffcount_matrix.tsv", sep='\t')
print(df)
ncar = 60
iter = int(float(len(line[1])/ncar))+1
#print(iter)

##-- Output to nexus format for network 
with open("haplo_matrix.nex",'w') as m:
    m.write("#nexus\n")
    m.write("begin taxa;\ndimensions ntax=74;\nend;\n")
    m.write("\n")
    m.write("begin characters;\ndimensions nchar="+str(len(line[1]))+";\nformat datatype=standard labels;\n")
    m.write("matrix \n")
    for i in range(iter):
        for k in seq.keys():
            m.write(("%s\t%s\n") % (k,seq[k][:ncar]))
            seq[k]=seq[k][ncar:]
        m.write("\n")
    m.write(";\n")
    m.write("end;")
    
```
